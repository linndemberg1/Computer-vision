```python
import numpy as np  # Importa a biblioteca NumPy para manipulação de arrays e operações matemáticas
import cv2  # Importa a biblioteca OpenCV para processamento de imagens
import matplotlib.pyplot as plt  # Importa a biblioteca Matplotlib para criação de gráficos e visualização

# Caminho da imagem a ser processada
image_path = '/content/vision.png'  # Caminho do arquivo da imagem em escala de cinza

# Carregar a imagem em escala de cinza
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)  # Lê a imagem do caminho especificado como uma imagem em escala de cinza

# Função para aplicar correção radiométrica à imagem
def radiometric_correction(image, a, b):
    # Corrige a imagem aplicando a transformação linear
    corrected_image = a * image + b  # Aplica a correção radiométrica
    corrected_image = np.clip(corrected_image, 0, 255)  # Garante que os valores estejam dentro do intervalo de 0 a 255
    return corrected_image.astype(np.uint8)  # Converte a imagem corrigida de volta para o tipo uint8

# Parâmetros da correção radiométrica
a = 1.2  # Fator de multiplicação
b = -30  # Valor de adição

# Aplica a correção radiométrica à imagem
corrected_image = radiometric_correction(image, a, b)

# Cria uma figura para exibir as imagens
plt.figure(figsize=(10, 5))  # Define o tamanho da figura

# Exibe a imagem original
plt.subplot(1, 2, 1)  # Define o local do primeiro sub-gráfico
plt.title('Imagem Original')  # Define o título do sub-gráfico
plt.imshow(image, cmap='gray')  # Exibe a imagem original em escala de cinza
plt.axis('off')  # Remove os eixos

# Exibe a imagem corrigida
plt.subplot(1, 2, 2)  # Define o local do segundo sub-gráfico
plt.title('Imagem Corrigida')  # Define o título do sub-gráfico
plt.imshow(corrected_image, cmap='gray')  # Exibe a imagem corrigida em escala de cinza
plt.axis('off')  # Remove os eixos

# Exibe a figura com as imagens
plt.show()  # Mostra a figura com as imagens lado a lado
```
